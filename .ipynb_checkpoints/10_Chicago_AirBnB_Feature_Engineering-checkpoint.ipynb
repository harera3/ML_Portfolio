## 1. The Data

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
pd.set_option("display.max_columns", 99)
pd.set_option("display.max_rows", 999)
pd.set_option('precision', 3)

chicago = pd.read_csv('data/chicago_airbnb2')

num_cols = ['host_rr', 'host_ar', 'latitude', 'longitude', 'accommodates',
            'baths', 'bedrooms', 'beds', 'min_nights', 'max_nights', 'n_reviews', 
            'rs_rating', 'rs_accuracy', 'rs_cleanliness', 'rs_check_in', 
            'rs_communication', 'rs_location', 'rs_value', 'reviews_per_month']

cat_cols = ['host_rt', 'superhost', 'neighborhood', 'property_type', 'room_type', 
            'instant_bookable']

chicago = chicago.drop('amenities', axis=1)

train, test = train_test_split(chicago, test_size=0.2, random_state=42)

X_train = train.drop('price', axis=1)
y_train = train['price']

X_test = test.drop('price', axis=1)
y_test = test['price']

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## 2. New Numerical Attributes via Custom Transformation

```python
from sklearn.base import BaseEstimator, TransformerMixin

class AttributeAdder(BaseEstimator, TransformerMixin):    
    def __init__(self):
        self    
        
    def fit(self, df, y = None):               
        return self

    def transform(self, df):        
        df_ = df.copy()
        df_['beds_per_bdrm'] = df_['beds']/df_['bedrooms']
        df_['baths_per_bed'] = df_['baths']/df_['beds']
        df_.replace(np.inf, 0, inplace=True)
        df_.replace(np.nan, 0, inplace=True)
        return df_
    
adder = AttributeAdder()
X_train_num = adder.fit_transform(train[num_cols])
print(X_train_num.shape)
X_train_num.head()
```

```python
num_cols = X_train_num.columns
num_cols
```

## 3. Scaling Numerical Attributes
### i. Before scaling:

```python
X_train_num.describe().T
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
ax = sns.boxplot(data=X_train_num, orient="h", palette="husl")
```

### ii. After Standardization:
We chose `StandardScaler` because it is less sensitive from outliers which we have in at least 3 columns (min_nights, max_nights, and n_reviews). I we didn't have outliers, we probably would have used `MinMaxScaler`.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_num)
X_train_scaled = pd.DataFrame(X_train_scaled, 
                             columns=num_cols)
X_train_scaled.describe().T
```

```python
plt.figure(figsize=(10,10))
ax = sns.boxplot(data=X_train_scaled, orient="h", palette="husl")
```

### iii. After Normalization:
Standadization doesn't seem that our data is scaled well. Let's try normalization and see the effect

```python
from sklearn.preprocessing import MinMaxScaler

normer = MinMaxScaler()
X_train_normed = normer.fit_transform(X_train_num)
X_train_normed = pd.DataFrame(X_train_normed, 
                             columns=num_cols)
X_train_normed.describe().T
```

```python
plt.figure(figsize=(10,10))
ax = sns.boxplot(data=X_train_normed, orient="h", palette="husl")
```

Normalization doesn't really look that much better than standardization, maybe we should have dealt with outliers better. We will use standardization during model training and see what we get. Maybe after narrowing our features to the best ones, we can go back and work on outliers and retrain the model and see if there are any improvements. 

For now, let's work on categorical attributes

## 4. Looking into Categorical Columns

```python
X_train_cat = train[cat_cols]
X_train_cat.head()
```

```python
X_train_cat.nunique()
```

```python
colors = ['blue', 'blue', 'red', 'brown', 'blue', 'blue']
X_train_cat.nunique().plot.bar(figsize=(12,6),
                              color=colors)
plt.title('Number of unique categories')
```

It looks like there are too many categories in `neighborhood`, and possibly `property_type`, that we need to look into. We may want to combine minor categories, especially in the `neighborhood` column, to avoid ending up with one-hot-encoded columns in the test set that are not in the train set. This is likely to happen especially if some categories are only represented by one value. Example: category `East Side` has only one value; if it ends up in the test set, out preprocessing pipeline `transform` function will fail because it will not have see the category in the train set where `fit` was done.
## 5. Combining Minor Categories

```python
chicago['neighborhood'].value_counts().nsmallest(25)
```

```python
print(X_train['neighborhood'].nunique())
print(X_test['neighborhood'].nunique())
```

Let's combine the smallest 24 categories (`value_count < 10`) into one category that we will call `Other`. Here is how:
    - We will convert the 24 smallest value counts to dictionary 
    - Then turn the dictionary's key into a list
    - Then replace every item in the list with the string `Other`
    - The new value counts shows that `Other`'s count is now `107`

```python
other_dict = chicago['neighborhood'].value_counts().nsmallest(24).to_dict()
other_list = list(other_dict.keys())
for item in other_list:
    chicago['neighborhood'] = chicago['neighborhood'].str.replace(item, 'Other')
    
chicago['neighborhood'].value_counts().head(20)
```

```python
print(X_train['property_type'].nunique())
print(X_test['property_type'].nunique())
```

```python
neibhd_freq = chicago['neighborhood'].value_counts(normalize=True)
fig = neibhd_freq.sort_values(ascending=False).plot.bar(figsize=(20,6))
fig.axhline(y=0.01, color='red')
fig.set_title('Identifying Rare Categories')
plt.show()
```

```python
prpty_freq = chicago['property_type'].value_counts(normalize=True)
fig = prpty_freq.sort_values(ascending=False).plot.bar(figsize=(20,6))
fig.axhline(y=0.05, color='red')
fig.set_title('Identifying Rare Categories')
plt.show()
```

It's hard to decide on combining `property_type` categories knowing that if we do we will lose some information on what actually makes the combined categories different. For example, combining `suite` with `bungalow` might actually hurt our model.

## 6. Encoding: Getting Dummies

```python
def make_dummies(df, column):
    dummies = pd.get_dummies(df[column], prefix=column)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop([column], axis=1)
    return df

train, test = train_test_split(chicago, test_size=0.2, random_state=42)
X_train = train.drop('price', axis=1)

dummy_cols = ['host_rt', 'neighborhood', 'property_type', 'room_type']
for col in dummy_cols:
    X_train = make_dummies(X_train, col)
    
print(X_train.shape)
X_train.head()
```

That's 93 columns. 

We used `get_dummies` here because we wanted to see the column it return. For model training, we will use `OneHotEncoding` which doesn't return dataframes but instead returns a numpy array. It also doesn't give us column names, which we won't need though. Let's end part 3 here and move on to model training!
